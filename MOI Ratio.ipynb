```python
# Dependencies and Setup
#importing functions into jupter notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#Creating a path to the data files
atlanta_home_data = "areas_clean_Reviewed_10.26.22.csv"

house_MOI_df = pd.read_csv(atlanta_home_data)

#display results
house_MOI_df.head()
```

```python
# add column for Month On Inventory, calculated as Homes Sold / Inventory
MOI = house_MOI_df["Homes Sold"]/house_MOI_df["Inventory"]
house_MOI_df["Months on Inventory"] = MOI

house_MOI_df.head()                                                    
```

```python
# Line plot MOI over time for each zip code using a for loop



# for each in regions:
#     house_MOI_df.plot('Month of Period End','MOI',kind='line')

# line plot for overall MOI over time
house_MOI_df.plot('Month of Period End', 'Months on Inventory', kind='line', figsize=(20,3))
```
