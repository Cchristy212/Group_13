```python
# Dependencies and Setup
#importing functions into jupter notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

#Creating a path to the data files

house_market_df = pd.read_csv("areas_clean.csv")

#display results
house_market_df.head()
```

Where should potential home buyers search for the most affordable home in several price ranges? (low, mid, high) (how much will you pay, any homes to buy, how quickly do you need to buy)


```python
house_market_df['Median Sale Price'] = house_market_df['Median Sale Price'].str.strip('$').str.replace(',','').astype('float')
```

```python
house_market_df.head()
```

```python
#group by Region
housing_by_region_mean = house_market_df.groupby('Region')['Median Sale Price'].mean()
housing_by_region_median =house_market_df.groupby('Region')['Median Sale Price'].median()
housing_by_region_variance =house_market_df.groupby('Region')['Median Sale Price'].var()
housing_by_region_stdev = house_market_df.groupby('Region')['Median Sale Price'].std()

housing_by_region_summary = pd.DataFrame({"Mean": housing_by_region_mean,'Median': housing_by_region_median, 
                                          'Variance': housing_by_region_variance,'Standard Deviation': housing_by_region_stdev})

housing_by_region_summary
```

```python
#inventory and median sale price
house_market_df.plot('Median Sale Price', 'Inventory',kind='scatter')
```

```python
#
house_market_df['year'] = pd.DatetimeIndex(house_market_df['Month of Period End']).year

# convert a series to list 
year = house_market_df['year'].tolist()
n = len(year)

#
year_list = []
counter = 0
for i in range(0,10): 
    year_list.append(house_market_df.loc[house_market_df['year'] == (2012 + i)])
year_list

#Display Results
house_market_df.head()
```

```python
year_list
```

```python
#
for year_seperation in year_list:
    width = 0.4
    data_ = pd.DataFrame(year_seperation)
    fig, ax = plt.subplots()
    ax.bar(x=data_['year'], height=data_['Median Sale Price']/max(data_['Inventory']), width=width, label='Median Sale Price')
    ax.bar(x=data_['year']+width, height=data_['Inventory'], width=width, label= 'Inventory')
    ax.set_xlabel(data_['year'].iat[0])
    ax.legend()
    
fig.savefig('year.png')
```

```python
width = 0.4

fig, ax = plt.subplots()
ax.bar(x=house_market_df['year'], height=house_market_df['Median Sale Price']/max(house_market_df['Median Sale Price']), width=width, label='Median Sale Price')
ax.bar(x=house_market_df['year']+width, height=house_market_df['Inventory']/max(house_market_df['Inventory']), width=width, label= 'Inventory')
ax.set_xlabel("year")
ax.legend()
```

```python
#Each Year with Inventory y axis and median sale price x axis

for year_seperation in year_list:

    scatter_plot_df = house_market_df.loc[house_market_df["Region"] == "Year",:]
    scatter_plot_df

    
    x_axis = scatter_plot_df["Median Sale Price"]
    y_axis = scatter_plot_df["Inventory"]


    fig1, ax1 = plt.subplots(figsize=(12, 8))
    plt.title(label=f"Year: {year_seperation}",loc="center")
    ax1.plot(x_axis, y_axis,linewidth=2, markersize=12,marker="o",color="blue", label="mm3")
    ax1.set_xlabel("Market Sale Price",fontsize =14)
    ax1.set_ylabel("Inventory",fontsize =14)

#Display Results
plt.show()
```
